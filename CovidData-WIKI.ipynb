```python
#!pip install bs4
# import Necessary Module
from urllib.request import urlopen
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
# getting url html document
try:
    #source = urlopen("https://www.google.com/covid19-map/").read()
    source= urlopen("https://en.m.wikipedia.org/wiki/Template:COVID-19_pandemic_data").read()
except Exception as e:
    print(e)
```

```python

%%time
# Converting to A BeautifulSoup object
try:
    soup= BeautifulSoup(source,"lxml")
    Cases= soup.find_all('tr')
except Exception as e:
    print(e)
```

```python

%%time

try:
    soup2= BeautifulSoup(source,'lxml')
    th= soup2.find_all('th')
except Exception as e:
    print(e)
th=th[0:492:2]
len(list(th))
```

```python

# Extracting Countries
countries=[]
for i in range(6,245):
    tex = th[i].text.replace("\n",'')
    countries.append(tex)
    
countries
```

```python
countries= np.array(countries).reshape(239,1)
countries
```

```python

df1 = pd.DataFrame(countries, columns=['Country'])
df1
```

```python
data =[]
for row in Cases:
    td=[td.string for td in row]
    data.append(td)
cols=data[2]
array = np.array(data)
print(array)

df3=pd.DataFrame(array)
df3
```

```python
ndata=[]
for row in range(2,237):
    
    s= str(array[row]).split()
    ndata.append(s[5])
    ndata.append(s[7])
    ndata.append(s[9])
    
newdata=[]
for data in ndata:
    val = data.replace("\\n","").replace("'","").strip()
    newdata.append(val)
df4= np.array(pd.DataFrame(newdata))
df4
```

```python
comma=[]
for row in df4:
    for i in row:
        v= str(i).replace(",","")
        comma.append(v)

resh= np.array(comma).reshape(235, 3)
print(resh)

    
```

```python
df=pd.DataFrame(resh, columns=['Confirmed','Deaths','Recovered'])
df.replace({'None':np.nan},inplace=True)
df
# Converting object to numeric data type
df1["Confirmed"]=(df["Confirmed"])
df1["Recovered"]=(df["Recovered"])
df1["Deaths"]=(df["Deaths"])

df1
```
